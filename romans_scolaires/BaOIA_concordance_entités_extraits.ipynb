```python

#@title Installation des bibliothèques nécessaires et connexion à un compte Google Drive.

''' 
Google Colab notebook.
Python == 3.7.11

BaOIA - La Contemporaine - Université de Nanterre
'''


## Installation des bibliothèques et connexion au compte Google Drive


import re
from google.colab import drive
import itertools
import json
import pandas as pd
import folium

drive.mount('/content/drive/')
%cd /content/drive/My Drive/romans_scolaires/
```

```python
fichier_json = "/content/drive/My Drive/romans_scolaires/test_textblocks.json"
with open(fichier_json, "r") as f :
  contenu_json = json.load(f)

nom_doc = []
lieu =[]
textblock = []
dict_excel = {}
contenu = []


for lieu in contenu_json:
  textblock = contenu_json[lieu]["Textblock"]
  nom_doc = contenu_json[lieu]["Document"]

  nom_open = "".join('/content/drive/My Drive/romans_scolaires/NER/' + nom_doc)
  with open(nom_open, "r") as opi:
      contenu_txt = json.load(opi)
      for text_block in contenu_txt["Text_Blocks"]:
          if textblock == text_block:
            contenu = contenu_txt["Text_Blocks"][text_block]["Content"]
            print(contenu)

    

      dict_excel[lieu] = {'Nom_doc': nom_doc, "Textblock": textblock, "Contenu": contenu}
with open('extrait_lieux.json', 'w', encoding='utf-8') as f:
    json.dump(dict_excel, f, ensure_ascii=False, indent=4)


df_contenu = pd.DataFrame.from_dict(dict_excel, orient = "index", columns=['Nom_doc', 'Textblock', "Contenu"])
print(df_contenu)
df_contenu.to_excel('extraits_des_lieux.xls', index = True, header=True)

```
