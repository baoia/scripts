```python
## Compter les personnages / lieux qui apparaissent le plus souvent dans les entités extraites

with open("merged_file.json",'r') as po:
  data_dict = json.load(po)
from collections import Counter

dictlist = []
for key, value in data_dict.items():
    dictlist.append(key)

with open("liste.txt", "w") as pu:
  pu.write(str(dictlist))
c1 = Counter(dictlist)

print(c1.most_common(4))

print(dictlist)
```

```python
## Récupération des noms des personnages dans liste pour compter le nombre de fois qu'ils apparaissent dans uen corpus

liste_persos = []

for file in glob.glob("/content/drive/My Drive/romans_scolaires/NER/*.json"):
  with open(file, "r") as ju:
    contenu_json = json.load(ju)

    for textblock_id in contenu_json["Text_Blocks"] :
      try: 
        for iji in contenu_json["Text_Blocks"][textblock_id]["NER-PERS"]:
          liste_persos.append(iji)
      except KeyError as ko:
        pass

c1 = Counter(liste_persos)

print(c1.most_common(120))

with open("liste_130_pers.txt", "w") as pi:
  pi.write(str(c1))
```

```python
## Idem lieux

liste_lieux = []

for file in glob.glob("/content/drive/My Drive/romans_scolaires/NER/*.json"):
  with open(file, "r") as ju2:
    contenu_json2 = json.load(ju2)

    for textblock_id2 in contenu_json2["Text_Blocks"] :
      try: 
        for iji2 in contenu_json2["Text_Blocks"][textblock_id2]["NER-LOC"]:
          liste_lieux.append(iji2)
      except KeyError as ko2:
        pass

c2 = Counter(liste_lieux)

print(c2.most_common(120))


with open("liste_130_lieux.txt", "w") as pi:
  pi.write(str(c2))
```
