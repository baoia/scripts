```
from openpyxl import load_workbook
import pandas as pd
import re
import requests
import shutil
import os
import xml.etree.ElementTree as ET
import lxml
from lxml import etree
from re import match
from bs4 import BeautifulSoup
import urllib.request, urllib.error, urllib.parse
import string
```

**Extraction et téléchargement des textes**

- Ligne 3: changer le nom de la page concernée du classeur xls. si nécessaire
- Ligne 4: changer l'identifiant de la colonne concernée par les liens ark si nécessaire
- Dernière ligne: indiquer le nom du classeur .xls. Ici par exemple, le fichier s'intitule "guides_paris.xlsx". 
- Lancer la cellule

```
def txt_gallica(fichier_xls):
  wb= load_workbook(fichier_xls)
  ws= wb['Paris']
  column_id = ws['D']
  id_list = [column_id[x].value for x in range(len(column_id))]
  print(id_list)
  for e in id_list[0:1000]:
    match = re.findall(r"b[a-z0-9]{4,}", e)
    print(match)

    for id in match:
        url2= "".join(["https://gallica.bnf.fr/ark:/12148/", id, '.texteBrut'])
        request = urllib.request.Request(url2)
        content = urllib.request.urlopen(request)

        parse = BeautifulSoup(content, 'html.parser')

        text1 = parse.find_all('body')
        filename = "".join([id, '2.txt'])
        text2 = re.sub('<.*?>', '', str(text1))
        text5 = re.sub("\d+", "", text2)
        with open(filename, 'w') as text_file:
            text_file.write(str(text5))


%cd /content/drive/My Drive/NER-demo/
txt_gallica(r"/content/drive/MyDrive/NER-demo/guides_paris.xlsx")
```
